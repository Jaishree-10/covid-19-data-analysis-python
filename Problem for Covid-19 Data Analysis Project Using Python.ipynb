```python
##Filter Required Columns
cols = ['continent','location','date','total_cases','total_deaths',
        'gdp_per_capita','human_development_index']
df=df[cols]
df.head()
```

```python
##Data Cleaning
##Remove Duplicates
df=df.drop_duplicates()
```

```python
##Find missing values
df.isnull().sum()
```

```python
###Remove rows where continent is missing
df=df.dropna(subset=["continent"])
```

```python
##Fill remaining missing values with 0
df=df.fillna(0)
```

```python
df.head(2)
```

```python
##Date Time Format
df["date"] = pd.to_datetime(df["date"])
```

```python
df.head(1)
```

```python
df["month"]=df["date"].dt.month
```

```python
df.head(3)
```

```python
##Data Aggregation
df_groupby=df.groupby("continent").max().reset_index()
##View new dataframe
df_groupby
```

```python
##Feature Engineering
df_groupby["total_deaths_to_total_cases"] = (
    df_groupby["total_deaths"] / df_groupby["total_cases"]
)
df_groupby["total_deaths_to_total_cases"]
```

```python
#Data Visualization
##Histogram of gdp_per_capita
sns.distplot(df["gdp_per_capita"])
plt.show()
```

```python
##Scatter plot — total_cases vs gdp_per_capita
sns.scatterplot(x=df["total_cases"] ,y=df["gdp_per_capita"])
plt.show()
```

```python
##Pairplot on df_groupby
sns.pairplot(df_groupby)
plt.show()
```

```python
##Bar plot — continent vs total_cases
sns.catplot(data=df_groupby, x="continent", y= "total_cases", kind="bar")
```
